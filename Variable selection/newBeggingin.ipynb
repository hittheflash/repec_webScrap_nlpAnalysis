```python
def nomu(nomus):
    long = nomus.split(" & ")
    b = long[1:-1]
    c = [i.split(" & ") for i in b]
    d = ["".join(i) for i in c]
    f = ";".join([long[0],long[-1]])
    g = f.split(";")
    return g
```

```python
def urlsJournals(journal):
    listaJ = []
    www = 'https://ideas.repec.org/a/' + journal
    response = get(www)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    for link in html_soup.find_all("a"):
        urr = link.get("href")
        parte1 = urr.split("y")
        if len(parte1)==2:
            parte2 = parte1[1].split("i")
            anio = int(parte2[0])
            if anio>2009:
                if urr.endswith("..html")==False:
                    listaJ.append(www + "/" + urr)
    return listaJ
```

```python
def writeLine(dirr):
    response = get(dirr)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    datas = []
    for link in html_soup.find_all("i"):
        repec = link.text
        if len(repec)>20:
            if repec[-1]=='.':
                repec = repec[:-1]
            parti = repec.split(":")
            year = parti[-5]
            vol = parti[4]
            issue = parti[-3]
            pages = parti[-1].split("-")
            pag0 = pages[0]
            pag1 = pages[1]
            data = [repec, year, vol, issue, pag0, pag1]
            if repec not in datas:
                datas+=data
    texto = ""
    for link in html_soup.find_all("script"):
        cadena = "Inner Text: {}".format(link.text)
        if "schema.org" in cadena:
            texto+=cadena
    part1 = texto.split('name": "')
    part2 = part1[1:]
    datos = []
    for i in part2:
        if "issueNumber" in i:
            part3 = i.split('",')
            journal = part3[0]
            datos.append(journal)
        elif "pagiurlssnation" in i:
            part4 = i.split('",')
            title = part4[0]
            autores = i.split('author": "')[1].split('",')[0]
            nombres = nomu(autores)
            datos.append(title)
            datos.append(nombres)
    L = []
    authrs = datos[2]
    for name in authrs:
        general = [datas[0]] + datos[:2][::-1] + datas[1:] + [name]
        linea = "|".join(general) + '\n'
        L.append(linea)
    return L
```

```python
def writeLineDescription(dirr):
    response = get(dirr)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    datas = []
    for link in html_soup.find_all("i"):
        repec = link.text
        if len(repec)>20:
            if repec[-1]=='.':
                repec = repec[:-1]
            parti = repec.split(":")
            year = parti[-5]
            vol = parti[4]
            issue = parti[-3]
            pages = parti[-1].split("-")
            pag0 = pages[0]
            pag1 = pages[1]
            data = [repec, year, vol, issue, pag0, pag1]
            if repec not in datas:
                datas+=data
    texto = ""
    for link in html_soup.find_all("script"):
        cadena = "Inner Text: {}".format(link.text)
        if "schema.org" in cadena:
            texto+=cadena
    part1 = texto.split('name": "')
    part2 = part1[1:]
    datos = []
    for i in part2:
        if "issueNumber" in i:
            part3 = i.split('",')
            journal = part3[0]
            datos.append(journal)
        elif "pagination" in i:
            part4 = i.split('",')
            title = part4[0]
            description = part4[2].split('": "')[1]
            datos.append(title)
            datos.append(description)
            
    descrip = datos[2]
    general = [datas[0]] + datos[:2][::-1] + datas[1:] + [descrip]
    linea = "|".join(general) + '\n'
    return linea
```
