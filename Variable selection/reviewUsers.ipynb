```python
file = open("authors_repec_by_journal.csv", "r")
linea = file.readline()
linea = file.readline()
while len(linea)!=0:
    t1 = time.time()
    idd = linea.split(";")[-1].strip()
    print(idd, type(idd))
    busqueda = " ".join(linea.split(";")[:-1]) + " " + idd + " repec"
    print(busqueda)
    urls = [j for j in search(busqueda, tld="co.in", num=10, stop=3, pause=2)]
    print(urls)
    TF = ["ideas" in i and idd in i for i in urls]
    print(TF)
    if sum(TF)>=1:
        print(linea.replace(";","\t"))
        file2.write(linea.replace(";","\t"))
        print("--------------------------------------------------------------------------------------")
    linea = file.readline()
    print((time.time()-t1)/60,"secs")
    print("--------------------------------------------------------------------------------------")
    print("--------------------------------------------------------------------------------------")
    print()
file.close()
file2.close()
```
