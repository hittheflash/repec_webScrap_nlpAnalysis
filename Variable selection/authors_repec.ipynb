```python
import json
import os
import requests 
from requests import get
from bs4 import BeautifulSoup
from googlesearch import search
import time
import sys
```

```python
# journal|year|author|repec
```

```python
csvs = [i for i in os.listdir() if i.endswith("v2_2.csv")]
csvs
```

```python
for csv in csvs:
    file1=open(csv, "r")
    nameRepec = csv.split("_v2_2")[0]
    file2=open("{}_repec_v3.csv".format(nameRepec), "w")
    file2.write("journal|year|author|repec\n")
    file3 = open("authors_error.txt", "a+")
    lineax = file1.readline()
    linea = file1.readline()
    while len(linea)!=0:
        datos = linea.strip().split("|")
        journal = datos[2]
        year = datos[3]
        author = datos[-1]
        # print(datos)
        #print(journal, year, author)
        
        busqueda = "{} {} repec".format(author,journal)
        try:
            print("Search 1:", busqueda)
            links = [j for j in search(busqueda, 
                                       tld="com",
                                       num=10,
                                       stop=5,
                                       pause=2) if "ideas.repec.org" in j and len(j.split("/"))==5 and len(j.split("/")[3])==1 and len(j.split("/")[-1].split(".")[0])>=4 and j.split("/")[-1].split(".")[0][:3].isalpha() and j.split("/")[-1].split(".")[0][3:].isdigit()]
            print(links)
            repecs = []
            if len(links)>0:
                for link in links:
                    response = get(link)
                    html_soup = BeautifulSoup(response.text, 'html.parser')

                    R = []

                    for link in html_soup.find_all("meta"):
                        forward = str(link.get("content"))
                        #print(forward)
                        if type(forward)!=None:
                            if "Short-ID" in forward:
                                R += [forward]
                                #print('forward',forward)

                    for link in html_soup.find_all("script"):
                        cadena = link.text
                        if len(cadena)>0 and '"@context": ' in cadena:
                            y = json.loads(cadena)
                            descripcion = y['description']
                            #print(descripcion)
                            if "Short-ID" in descripcion:
                                repec = descripcion.split('Short-ID ')[-1]
                                if repec not in repecs:
                                    #print('repec',repec)
                                    R += [repec]
                    print("Info by the name:", R[0])
                    if len(R)==2:
                        repecs.append(R)
            #print(repecs)
            else:
                line = "{}|{}|{}|none\n".format(journal, year, author)
                file2.write(line)
                print(line
                

            if len(repecs)>0:

                for r in repecs:
                    print("Info by the repec:", r[0])
                    #time.sleep(3)
                    busqueda = "{} repec".format(r[-1])
                    print("Search 2:", busqueda)
                    links = [j for j in search(busqueda, 
                                               tld="com",
                                               num=10,
                                               stop=5,
                                               pause=2) if "ideas.repec.org" in j and len(j.split("/"))==5 and len(j.split("/")[3])==1 and len(j.split("/")[-1].split(".")[0])>=4 and j.split("/")[-1].split(".")[0][:3].isalpha() and j.split("/")[-1].split(".")[0][3:].isdigit()]
                    for link in links:
                        response = get(links[0])
                        html_soup = BeautifulSoup(response.text, 'html.parser')

                        for link in html_soup.find_all("meta"):
                            forward = str(link.get("content"))
                            if type(forward)!=None:
                                if "Short-ID" in forward:
                                    if forward==r[0]:
                                        line = "{}|{}|{}|{}\n".format(journal, year, author, repec)
                                        file2.write(line)
                                        print(line)
                                    else:
                                        line = "{}|{}|{}|none\n".format(journal, year, author)
                                        file2.write(line)
                                        print(line)
        except ConnectionError:
            print("CONNECTION ERROR")
        except:
            print("Unexpected error:", sys.exc_info()[0])
            print("Revisar usuario:", author)
                      file3.write("{},{}".format(author, journal))
        linea = file1.readline()
        print()
        
    file1.close()
    file2.close()
    file3.close()
    print("--------------------------------------------")
```
