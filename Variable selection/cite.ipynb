```python
import time
import requests 
from requests import get
from bs4 import BeautifulSoup
```

```python
response = get('http://citec.repec.org/RePEc:eee:jfpoli:v:85:y:2019:i:c:p:72-78')
html_soup = BeautifulSoup(response.text, 'html.parser')
html_soup
```

```python
L = [0,0,0]
for link in html_soup.find_all("a"):
    cite = link.text
    if "Cited" in cite and "times" in cite:
        part = int(link.text.split(" ")[1])
        L[0]+=part
    elif "Cites" in cite and "references" in cite:
        part = int(link.text.split(" ")[1])
        L[1]+=part
    elif "Cocites" in cite and "documents" in cite:
        part = int(link.text.split(" ")[1])
        L[2]+=part
```

```python
L
```

```python
for link in html_soup.find_all("li"):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
    print("class: {}".format(link.get("class")))
    print("---------------------------------------------------------------------------------------------------------")
```

```python
C = []
for item in html_soup.find_all('li',class_=["Cell3Font",""]):
    print("{}".format(item.text))
    C.append(item.text)
    print("--------------------------------------------------------------------------------------------------------")
```

```python
C = C[L[0]:L[1]+1]
C
```
