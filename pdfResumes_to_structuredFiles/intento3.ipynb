```python
from pdfminer.pdfinterp import PDFResourceManager, process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from io import StringIO

def convert_pdf(path):

    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)

    fp = open(path, 'rb')
    process_pdf(rsrcmgr, device, fp)
    fp.close()
    device.close()

    strr = retstr.getvalue()
    retstr.close()
    return strr
```

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.converter import XMLConverter, HTMLConverter, TextConverter
from pdfminer.layout import LAParams
import io

def pdfparser(pdf):

    fp = open(pdf, 'rb')
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    # Create a PDF interpreter object.
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    # Process each page contained in the document.
    
    pgs = 0

    for page in PDFPage.get_pages(fp):
        interpreter.process_page(page)
        data =  retstr.getvalue()
        pgs+=1
    
    infos =  data.split("\n")
    P = []
    for i in infos:
        Y = i.split("\uf0b7")
        for j in Y:
            P.append(j)
    infos = P
    print(len(infos))
    infos = [i.replace("\x0c", "") for i in infos]
    print(len(infos))
    infos = [i.replace("\'\'", "") for i in infos]
    print(len(infos))
    infos = [i for i in infos if i!='']
    print(len(infos))
    infos = [i for i in infos if i!=" "]
    print(len(infos))
    infos = [i for i in infos if i!="  "]
    print(len(infos))
    infos = [str(i) for i in infos if i not in [str(i) for i in range(1, pgs+1)]]
    print(len(infos))
    #return infos

    n = len(infos)
    cvs = []
    for i in range(len(infos)):
        linea = infos[i]
        j = i+1
        if j<n:
            forward = infos[j]
            cond1 = "(" in linea and ")" not in linea
            cond2 = linea.count("(") > linea.count(")")
            cond3 = linea.endswith(", ")
            cond4 = linea.endswith("of ")
            cond5 = linea.endswith(": ")
            cond6 = linea.endswith(":")
            if cond1 or cond2 or cond3 or cond4 or cond5 or cond6:
                L = linea+forward
                cvs.append(L)
                    
    cvs = [i for i in cvs if i[0]!="("]
    print(len(cvs))
    return data, cvs
```
