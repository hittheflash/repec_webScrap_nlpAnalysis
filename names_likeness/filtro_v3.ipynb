```python
trues = []   #nombre con coincidencias
falses = []  # nombre sin coincidencias
m_igual = []
```

```python
G = open("prob75.txt", "r")
linea = G.readline()
while len(linea)!=0:
    part = linea.strip().split(" | ")
    x, y, a, b = part
    TF = get_bool_name(x, y)
    if float(b)>0.91:
        m_igual.append([x, y, a, b])
    elif TF==True:
        trues.append([x, y, a, b])
    else:
        falses.append([x, y, a, b])
    linea = G.readline()
G.close()
```

### Se procede a merge-ar los DataFrames

```python
import pandas as pd
import numpy as np
import os
```

```python
df_eco= pd.read_csv('emetrp.csv', delimiter=';',encoding = "ISO-8859-1", error_bad_lines=False)
df_aer= pd.read_csv('aecrev_v2.csv', delimiter=';',encoding = "ISO-8859-1", error_bad_lines=False)
df_jpe= pd.read_csv('jpolec.csv', delimiter=';',encoding = "ISO-8859-1", error_bad_lines=False)
df_qje= pd.read_csv('qjecon.csv', delimiter=';',encoding = "ISO-8859-1", error_bad_lines=False)
df_res= pd.read_csv('restud.csv', delimiter=';',encoding = "ISO-8859-1", error_bad_lines=False)
```

```python
df=df_aer.append(df_eco,ignore_index=True)
df=df.append(df_qje,ignore_index=True)
df=df.append(df_jpe,ignore_index=True)
df=df.append(df_res,ignore_index=True)
```

```python
df.drop(columns=['handle','title','vol','issue','page','end page'], inplace=True)
```

```python
df
```

```python
df_repec=pd.read_csv('autores_repec.csv', delimiter=',')
df_repec
```

```python
df_final = pd.merge(df, df_repec, how = 'left',left_on ='author', right_on = 'Name')
df_final
```

### Primer DataFrame creado con las coincidencias
##### No se requirio ningun filtro

```python
df_final_1 = df_final.dropna()
df_final_1
```

### Segundo Dataframe con valores de repec no encontrados
#### Se procede a procesar usando un filtro de comparacion

```python
df_final_0 = df_final[df_final.isna().any(axis=1)]
df_final_0 = df_final_0[['journal', 'year', 'author']]
df_final_0.sort_values(by='author')
```

```python
autors_ = list(df_final_0["author"])
lista = []
for i in autors_:
    if type(i)==float:
        lista.append(i)
    else:
        lista.append(i.lower())
df_final_0['author'] = lista
```
