```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
def comparar2(cadena1, cadena2):
    
    cadena1 = cadena1.replace(".", ". ").replace("  ", " ").strip().replace(",", "")
    cadena2 = cadena2.replace(".", ". ").replace("  ", " ").strip().replace(",", "")

    l1, l2 = cadena1.split(" "), cadena2.split(" ")
    if len(cadena1)<len(cadena2):
        fd = l1
        l1 = l2
        l2 = fd
    
    if len(l1)==len(l2) and len(l1)>1:
        verdad = []
        unos = []
        for i in range(len(l1)):
            lett1 = l1[i][0]
            lett2 = l2[i][0]
            uno = similar(l1[i], l2[i])
            if lett1==lett2:
                verdad.append(True)
            if uno==1.0:
                unos.append(uno)
                
        if sum(verdad)==len(l1):
            if len(l1)==2 and sum(unos)>=1.8:
                return True
            elif len(l1)>2 and sum(unos)>=len(l1)-1:
                return True
            else:
                return False
        else:
            return False
            
            
    elif len(l1)==3 and len(l2)==2 or len(l1)==2 and len(l2)==3:
        lett1_1 = l1[0][0]
        lett2_1 = l2[0][0]

        lett1_2 = l1[-1][0]
        lett2_2 = l2[-1][0]
        
        verdad = [0, 0]
        
        indicador = 0
        for i in l1:
            for j in l2:
                nn = similar(i, j)
                if nn>=0.8:
                    indicador+=nn
        if lett1_1==lett2_1 and lett1_2==lett2_2 and indicador >= len(l1)-1:
            return True
        else:
            return False
        
        
    elif len(l1)>3 and len(l2)>=3 or len(l1)>=3 and len(l2)>3:

            
        indicador = 0
        for i in l1:
            for j in l2:
                nn = similar(i, j)
                if nn>=0.8:
                    indicador+=nn
        if indicador >= len(l1)-1:
            return True
        else:
            return False
    else:
        return False
```

```python
trues = []   #nombre con coincidencias
falses = []  # nombre sin coincidencias
```

```python
G = open("prob75.txt", "r")
linea = G.readline()
while len(linea)!=0:
    part = linea.strip().split(" | ")
    x, y, a, b = part
    TF = comparar2(x, y)
    if TF==True or float(b)>0.91:
        trues.append([x, y, a, b])
    else:
        falses.append([x, y, a, b])
    linea = G.readline()
G.close()
```
