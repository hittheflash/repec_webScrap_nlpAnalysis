```python
def retornn(lista):
    rr = []
    for i in lista:
        if "pro" not in i:
            repec = i.split("/")[-1].split(".")[0]
            rr.append(repec)
        else:
            repec = i.split("/")[-2]
            rr.append(repec)
    
    if len(rr)==2:
        if rr[0]==rr[-1]:
            return rr[0]
        else:
            return 0
    else:
        return 0
```
