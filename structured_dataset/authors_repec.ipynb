```python
import os
import string
import pandas as pd
```

```python
letras = string.ascii_uppercase
total = 0
for i in range(len(letras)):
    archivos = os.listdir(letras[i])
    print("{}: {}".format(letras[i], len(archivos)), end="|")
    total+=len(archivos)
```

```python
total
```

```python
formatos = set()
for letra in letras:
    print(letra, end="|")
    archivos = os.listdir(letra)
    for archivo in archivos:
        F = open("{}/{}".format(letra, archivo), "r", encoding="utf8")
        linea = F.readline()
        while len(linea)!=0:
            if ":" in linea:
                titulo = linea.split(":")[0]
                formatos.add(titulo)
            linea = F.readline()
        F.close()
```

```python
formatos
```

```python
AUTHORS = []
for letra in letras:
    print(letra, end="|")
    archivos = os.listdir(letra)
    for archivo in archivos:
        dirr = "{}/{}".format(letra, archivo)
        F = open(dirr, "r", encoding="utf8")
        cadena = "\n".join(F.readlines())
        if 'Name-ASCII' in cadena:
            part = cadena.split("Name-ASCII: ")[-1]
            ascci = part.split("\n")[0]
            part2 = cadena.split("Short-Id: ")[-1]
            repec = part2.split("\n")[0]
            AUTHORS.append([ascci.title(), repec])
        else:
            part = cadena.split("Name-Full: ")[-1]
            full = part.split("\n")[0]
            part2 = cadena.split("Short-Id: ")[-1]
            repec = part2.split("\n")[0]
            AUTHORS.append([full.title(), repec])
        F.close()
```

```python
df = pd.DataFrame(AUTHORS)
df.columns = ["Name", "Short-Id"]
df
```

```python
export_csv = df.to_csv ('autores_repec.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```
